# Decision Tree Model On Scholarship Allocation 
### Based on Completion rate of education

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
edu= pd.read_csv('Completion rate of education.csv')
edu.head()
```

```python
#Make a copy of the dataset
edu1 = edu.copy()
```

```python
# Add a new column 'Scholarship' with 1 for values in 'Value' >= 50, 0 otherwise
edu1['Scholarship'] = edu1['Value'].apply(lambda x: 1 if x >= 50 else 0)
edu1.head()
```

```python
edu1.drop('Metadata', axis=1, inplace=True)
```

```python
#Declare the feature and target
inputs = edu1.drop('Scholarship', axis ='columns')
target =edu1['Scholarship']
```

```python
inputs
```

```python
target
```

```python
# machine learning works better with numeric value, transform with LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
le_region = LabelEncoder()
le_country = LabelEncoder()
le_indicator_name = LabelEncoder()

inputs['Region_n'] = le_region.fit_transform(inputs['Region'])
inputs['Country_n'] = le_country.fit_transform(inputs['Country'])
inputs['Indicator_Name_n'] = le_indicator_name.fit_transform(inputs['Indicator Name'])
```

```python
inputs.head()
```

```python
#Drop columns not used
inputs_n =inputs.drop(['Region', 'Country', 'Indicator Name','Target','Indicator Number'], axis= 'columns')
inputs_n
```

```python
#import decision tree library
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
#train the model
model.fit(inputs_n,target)
```

```python
#check the model score
model.score(inputs_n,target)
```

```python
# Predict to see the output
model.predict([[2030,20,3,1,1]])
```
